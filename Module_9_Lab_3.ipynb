Central to CNNs, a convolution operation is a linear element-wise multiplication operation between a small filter/kernel and same-sized patch from the image. We move this filter over the image like a sliding window from top left to bottom right. For each point on the image, a value is calculated based on the filter using a convolution operation. These filters can do simplest task like checking if there is a vertical line in the image or complicated task like detecting a human eye in the image.

Let's look at the convolution formula:

Convolution between image
$f(x, y)$ and kernel $k(x, y)$ is
$$f(x,y) * k(x,y) = \sum \limits _{i=0} ^{W-1} \sum \limits _{j=0} ^{H-1} f(i, j) k(x − i, y − j)$$

where $W$ and $H$ are the the width and height of the image.

```python
# 2D 3x3 binary image with vertical edge
image1 = np.array([[1,1,0], [1,1,0], [1,1,0]])

# 2D 3x3 binary image with horizontal edge
image2 = np.array([[0,0,0], [0,0,0], [1,1,1]])

# On plotting the images
plt.imshow(image1, cmap='gray', extent=[0, 3, 3, 0])
plt.show()
plt.imshow(image2, cmap='gray', extent=[0, 3, 3, 0])
plt.show()
```

```python
# Vertical Line filter
filter = np.array([[1,0,-1],
                   [1,0,-1],
                   [1,0,-1]])

# Applying filter to first image
output = np.sum(np.multiply(image1, filter))
print('Output from first image: ', output)

# Applying filter to second image
output = np.sum(np.multiply(image2, filter))
print('Output from second image: ', output)
```

```python
# Horizontal edge filter
filter = np.array([[-1,-1,-1],
                   [ 0, 0, 0],
                   [ 1, 1, 1]])

output = np.sum(np.multiply(image1, filter))
print('Output from first image: ', output)

output = np.sum(np.multiply(image2, filter))
print('Output from second image: ', output)
```

Non-zero output suggests that there is a vertical edge present in the first image and not present in the second image. Similarly, horizontal edge is detected in second.

Let's define a function to use convolution layer from Pytorch and use our own kernel to detect edges in image

```python
def apply_conv(image, kernel, padding=0, stride=1):

  #--------IMAGE PREPROCESSING-------
  image = torch.from_numpy(image)
  # Pytorch requires input to convolution in (N,C,H,W), where N = batch size and C=#channels in input
  input = image.view((1,1,image.shape[0], image.shape[1]))

  # --------------KERNEL-------------
  kernel = torch.from_numpy(kernel.astype(np.float32))

  # Pytorch requires kernel of shape (N,C,H,W), where N = batch size and C=#channels in input
  kernel = kernel.view((1,1,kernel.shape[0], kernel.shape[1]))

  # ---------CONVOLUTION LAYER from Pytorch--------
  conv = Conv2d(in_channels=1, out_channels=1, kernel_size=kernel.shape, padding=padding, stride=stride)

  # Set the kernel weights in the convolution layer
  conv.weight = torch.nn.Parameter(kernel)

  # ---------APPLY CONVOLUTION--------
  output = conv(input.float())
  output_img = output.data.numpy()  # Tensor to back in numpy
  output_img = output_img.reshape((-1, output_img.shape[-1])) # Reshape to 2D image

  return output_img
```

## Softmax layer/activation
Recall that logistic regression produces a decimal between 0 and 1.0. For example, a logistic regression output of 0.8 from an email classifier suggests an 80% chance of an email being spam and a 20% chance of it being not spam. Clearly, the sum of the probabilities of an email being either spam or not spam is 1.0.

Softmax extends this idea into a multi-class world. That is, Softmax assigns decimal probabilities to each class in a multi-class problem. Those decimal probabilities must add up to 1.0. This additional constraint helps training converge more quickly than it otherwise would.
Softmax is implemented through a neural network layer just before the output layer. The Softmax layer must have the same number of nodes as the output layer.

## Learning to train a CNN network

```python
# Import packages
import torch
import torchvision
import torchvision.transforms as transforms
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import matplotlib.pyplot as plt
import numpy as np
```

```python
#  Images returned from torchvision dataset classes is in range [0,1]
# We transform them to tensors and normalize them to range [-1,1] using 'Normalize' transform

transform = transforms.Compose(
    [transforms.ToTensor(),
     transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))])

batch_size = 128

trainset = torchvision.datasets.CIFAR10(root='./data', train=True,
                                        download=True, transform=transform)
trainloader = torch.utils.data.DataLoader(trainset, batch_size=batch_size,
                                          shuffle=True, num_workers=2)

testset = torchvision.datasets.CIFAR10(root='./data', train=False,
                                       download=True, transform=transform)
testloader = torch.utils.data.DataLoader(testset, batch_size=batch_size,
                                         shuffle=False, num_workers=2)

# Classes in CIFAR10
classes = ('plane', 'car', 'bird', 'cat',
           'deer', 'dog', 'frog', 'horse', 'ship', 'truck')
```

```python
print('Training data shape : ', trainset.data.shape, len(trainset.targets))
print('Testing data shape : ', testset.data.shape, len(testset.targets))

# Find the unique numbers from the train labels
nClasses = len(classes)
print('Total number of outputs : ', nClasses)
print('Output classes : ', classes)
```

```python
def train(num_epochs, model, train_loader, loss_func, optimizer):

  # Training mode
  model.train()

  train_losses = []
  train_acc = []

  # Train the model
  for epoch in range(num_epochs):
    running_loss = 0
    running_acc = 0
    for i, (images, labels) in enumerate(train_loader):

      # clear gradients for this training step
      optimizer.zero_grad()

      # Forward pass
      output = model(images)

      # Calculate loss
      loss = loss_func(output, labels)

      # Backpropagation, compute gradients
      loss.backward()

      # Apply gradients
      optimizer.step()

      # Running loss
      running_loss += loss.item()

      # indices of max probabilities
      _, preds = torch.max(output, dim=1)

      # Calculate number of correct predictions
      correct = (preds.float() == labels).sum()
      running_acc += correct

      # Average loss and acc values
      epoch_loss = running_loss / len(train_loader.dataset)
      epoch_acc = running_acc / len(train_loader.dataset)

    train_losses.append(epoch_loss)
    train_acc.append(epoch_acc)
    print ('Epoch {}/{}, Loss: {:.4f}, Accuracy: {:.4f}'.format(epoch + 1, num_epochs, epoch_loss, epoch_acc*100))

  return train_losses, train_acc
```

```python
def test_model(model, testloader):
  correct = 0
  total = 0
  # Deactivate autograd engine (don't compute grads since we're not training)
  with torch.no_grad():
    for data in testloader:
        images, labels = data
        # Calculate outputs by running images through the network
        outputs = model(images)
        # The class with the highest value is what we choose as prediction
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

  print('Accuracy of the network: %d %%' % (
      100 * correct / total))
```

```python
# CNN with 2 CONV layers and 3 FC layers
class Net(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = nn.Conv2d(3, 16, 3)
        self.pool = nn.MaxPool2d(2, 2)
        self.conv2 = nn.Conv2d(16, 32, 5)
        self.fc1 = nn.Linear(32 * 5 * 5, 512)
        self.fc2 = nn.Linear(512, 128)
        # output layer 10 classes
        self.fc3 = nn.Linear(128, 10)

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        # flatten all dimensions except batch
        x = torch.flatten(x, 1)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x
```

```python
model = Net()
print(model)
```

```python
# Cross Entropy loss for multi-class classification
criterion = nn.CrossEntropyLoss()
```

```python
# SGD optimizer with momentum
optimizer = optim.SGD(model.parameters(), lr=0.05, momentum=0.9)
```

```python
num_epochs = 5  # iterations
train_losses, train_acc = train(num_epochs, model, trainloader, criterion, optimizer)
```

```python
fig = plt.figure(figsize=(10,4))
ax = fig.add_subplot(1,2, 1)
ax.plot(np.arange(1,len(train_losses)+1),train_losses)
plt.xlabel('Training loss')
plt.ylabel('Epochs')
ax.set_title('Loss vs Epochs')
ax = fig.add_subplot(1,2, 2)
ax.plot(np.arange(1,len(train_acc)+1),train_acc)
plt.xlabel('Training accuracy')
plt.ylabel('Epochs')
ax.set_title('Accuracy vs Epochs')
plt.show()
```

```python
# Accuracy on test data after training
test_model(model, testloader)
```

Convolutional Neural Networks (CNNs) are generally preferred over simple Artificial Neural Networks (ANNs) for image classification tasks due to several reasons:


Spatial Hierarchical Structure: CNNs are specifically designed to capture spatial hierarchies in images. They use convolutional layers that detect low-level features (such as edges and textures) and progressively combine them to detect higher-level features (such as shapes and objects). This hierarchical structure helps CNNs learn more meaningful representations of images compared to ANNs, which treat input features as independent.

Parameter Sharing: CNNs exploit parameter sharing, where the same set of weights is applied across different spatial locations in the input image. This reduces the number of parameters in the model and helps capture translational invariance. In contrast, ANNs have a fully connected architecture where each neuron in one layer is connected to every neuron in the subsequent layer, resulting in a much larger number of parameters and a lack of translational invariance.


Local Connectivity: CNNs leverage local connectivity by connecting each neuron to a local region of the input volume. This local connectivity allows CNNs to focus on local patterns and relationships in the image, making them more suitable for tasks like image classification where spatial information is crucial. ANNs, on the other hand, treat all input features equally and do not consider spatial relationships explicitly.


Pooling Layers: CNNs typically include pooling layers (such as max pooling or average pooling) after convolutional layers, which downsample the feature maps and reduce their dimensionality while preserving important information. Pooling helps make the learned features more robust to small variations in the input and reduces the computational complexity of the network. ANNs do not have built-in mechanisms for spatial downsampling and may struggle with large input sizes.


Pretrained Models and Transfer Learning: CNNs have been extensively pretrained on large image datasets (e.g., ImageNet), which allows researchers and practitioners to leverage prelearned features for a wide range of image classification tasks through transfer learning. Fine-tuning pretrained CNNs on task-specific datasets often leads to faster convergence and better generalization compared to training ANNs from scratch.
Efficiency: Due to their architectural design and parameter sharing, CNNs are generally more computationally efficient than ANNs for image-related tasks. They can process large images with relatively few parameters compared to fully connected ANNs, making them more practical for real-world applications, especially in scenarios with limited computational resources.

Reducing the number of convolutional layers to only one in a CNN can have several implications:


Decreased Capacity for Feature Extraction: Convolutional layers are responsible for extracting features from the input image by applying filters (kernels) across the input spatial dimensions. With only one convolutional layer, the network has limited capacity to extract diverse and hierarchical features from the image. This may result in the model being unable to capture complex patterns and structures in the data, leading to lower performance.


Loss of Hierarchical Representation Learning: CNNs with multiple convolutional layers are able to learn hierarchical representations of the input data. Each convolutional layer captures increasingly abstract features, starting from low-level features (e.g., edges, textures) in the early layers to high-level semantic features (e.g., object parts, object shapes) in the later layers. With only one convolutional layer, the network may struggle to learn such hierarchical representations, limiting its ability to understand and generalize from the input images.


Reduced Model Capacity: CNNs with multiple convolutional layers typically have larger model capacity, allowing them to learn more complex relationships in the data. By reducing the number of convolutional layers to one, the model's capacity is significantly reduced, potentially leading to underfitting, where the model is unable to capture the underlying patterns in the data.


Increased Sensitivity to Position and Scale: Multiple convolutional layers with pooling operations help CNNs achieve translation and scale invariance, making them robust to variations in object position and size within the input image. With only one convolutional layer, the network may be more sensitive to changes in position and scale, making it less robust to such variations and potentially leading to lower generalization performance.

In traditional CNN architectures for image classification tasks, the softmax activation function is typically used in the output layer to transform the raw output scores of the network into probability values representing the likelihood of each class. However, it's worth noting that the softmax activation function is often implicitly included as part of the cross-entropy loss function used for training the CNN.


Here are a few reasons why the softmax activation function is not explicitly used in the last layer of a CNN:


Cross-Entropy Loss: The softmax activation function is commonly used in conjunction with the categorical cross-entropy loss function. During training, the softmax function is applied to the output logits of the network before computing the cross-entropy loss. This combined softmax-cross-entropy operation effectively encourages the network to produce output probability distributions that sum to one and are more suitable for classification tasks.


Numerical Stability: When training deep neural networks, especially CNNs with multiple layers, numerical stability becomes crucial. Directly applying the softmax activation function to the output logits of the last layer may lead to numerical instability, especially when dealing with large numbers. By incorporating the softmax operation as part of the cross-entropy loss calculation, numerical stability can be maintained more effectively.


Efficiency: Separating the softmax activation function from the last layer of the network allows for greater flexibility in network design and optimization. For example, in some cases, researchers may choose to use alternative output activation functions (e.g., linear activation for regression tasks) or custom activation functions tailored to specific requirements. By decoupling the activation function from the last layer, CNN architectures can be more versatile and adaptable to different tasks.


Gradient Flow: During backpropagation, the gradient of the cross-entropy loss with respect to the logits is computed directly, simplifying the gradient flow through the network. This can lead to more stable and efficient training compared to networks where the softmax activation is included as part of the last layer.
